```python
# Group by Province and calculate the total Area
grouped_data = df1.groupby('Country / Province', as_index=False)['Area \n(km2)'].sum()

# Create a bar chart using the aggregated data
plt.figure(figsize=(10, 6))
bars = plt.bar(grouped_data['Country / Province'], grouped_data['Area \n(km2)'], color='lightgreen')

# Add numeric values on top of each bar
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # X-coordinate (center of the bar)
        bar.get_height(),                  # Y-coordinate (height of the bar)
        f'{int(bar.get_height()):,}',      # Text (formatted with commas for large numbers)
        ha='center', va='bottom', fontsize=10  # Text alignment
    )

# Disable scientific notation on y-axis
plt.ticklabel_format(style='plain', axis='y')  # Prevent scientific notation

# Add title and labels
plt.title('Total Area by Province', fontsize=16)
plt.xlabel('Province', fontsize=14)
plt.ylabel('Area (in sq. km)', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Show the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define high-risk threshold for floodwater extent
high_risk_threshold = 1000  # Adjust threshold based on your dataset

# Filter only high-risk provinces
high_risk_data = df1[df1['Maximum flood \nwater extent \n[24 September - 07 October 2024] (km2)'] > high_risk_threshold]

# Scale down bubble sizes for better visibility
bubble_size = high_risk_data['Maximum flood \nwater extent \n[24 September - 07 October 2024] (km2)'] * 0.3  # Adjust scaling factor

# Define colors for high-risk provinces
colors = ['red' for _ in high_risk_data['Maximum flood \nwater extent \n[24 September - 07 October 2024] (km2)']]

# Create a figure
plt.figure(figsize=(12, 6))

# Scatter plot for high-risk provinces
plt.scatter(
    high_risk_data['Country / Province'],
    high_risk_data['Maximum flood \nwater extent \n[24 September - 07 October 2024] (km2)'],
    s=bubble_size,
    c=colors,
    alpha=0.7,
    edgecolors='black'
)

# Annotate each province
for i, province in enumerate(high_risk_data['Country / Province']):
    extent = high_risk_data['Maximum flood \nwater extent \n[24 September - 07 October 2024] (km2)'].iloc[i]
    plt.text(
        province, extent + 50,  # Offset for clarity
        province,
        fontsize=10, ha='center', va='bottom', color='black'
    )

# Add titles and labels
plt.title('High-Risk Provinces in Pakistan Based on Flood Extent (24 Sept - 07 Oct 2024)', fontsize=16)
plt.xlabel('Province', fontsize=14)
plt.ylabel('Flood Water Extent (km²)', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define high-risk threshold for floodwater extent
high_risk_threshold = 100  # Adjust threshold based on your dataset

# Filter only high-risk districts
high_risk_data = df2[df2['Maximum flood \nwater extent \n[24 September - 07 October 2024] (km2)'] > high_risk_threshold]

# Scale down bubble sizes for better visibility
bubble_size = high_risk_data['Maximum flood \nwater extent \n[24 September - 07 October 2024] (km2)'] * 0.3  # Adjust scaling factor

# Define colors for high-risk districts
colors = ['red' for _ in high_risk_data['Maximum flood \nwater extent \n[24 September - 07 October 2024] (km2)']]

# Create a figure
plt.figure(figsize=(12, 8))  # Increase figure size for better visibility

# Scatter plot for high-risk districts
plt.scatter(
    high_risk_data['District'],  # Use 'District' for the x-axis
    high_risk_data['Maximum flood \nwater extent \n[24 September - 07 October 2024] (km2)'],
    s=bubble_size,
    c=colors,
    alpha=0.7,
    edgecolors='black'
)

# Annotate each district
for i, district in enumerate(high_risk_data['District']):  # Use 'District' for annotations
    extent = high_risk_data['Maximum flood \nwater extent \n[24 September - 07 October 2024] (km2)'].iloc[i]
    plt.text(
        district, extent + 50,  # Offset for clarity
        district,
        fontsize=10, ha='center', va='bottom', color='black'
    )

# Add titles and labels
plt.title('High-Risk Districts of Balochistan in Pakistan Based on Flood Extent (24 Sept - 07 Oct 2024)', fontsize=16)
plt.xlabel('District', fontsize=14)
plt.ylabel('Flood Water Extent (km²)', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Adjust y-axis limits to add padding above the highest bubble
max_extent = high_risk_data['Maximum flood \nwater extent \n[24 September - 07 October 2024] (km2)'].max()
plt.ylim(0, max_extent + 200)  # Add padding of 200 km² above the maximum extent

# Add grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define threshold for population exposed to maximum flood water extent
population_threshold = 1000 # Adjust threshold as needed

# Filter data for districts with population exposed above the threshold
high_risk_data = df2[df2['Population potentially exposed to maximum flood water extent\n[24 September - 07 October 2024]'] > population_threshold]

# Create a figure with an appropriate size
plt.figure(figsize=(14, 8))

# Line plot for Population potentially exposed to maximum flood water extent
plt.plot(
    high_risk_data['District'],
    high_risk_data['Population potentially exposed to maximum flood water extent\n[24 September - 07 October 2024]'],
    marker='o',
    linestyle='-',
    color='b',
    label='Population Exposed to Max Flood'
)

# Add titles and labels
plt.title('Population Potentially Exposed to Maximum Flood Water Extent (24 Sept - 07 Oct 2024)', fontsize=16)
plt.xlabel('District', fontsize=14)
plt.ylabel('Population Exposed to Maximum Flood', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Annotate each point with the value
for i, value in enumerate(high_risk_data['Population potentially exposed to maximum flood water extent\n[24 September - 07 October 2024]']):
    plt.text(
        high_risk_data['District'].iloc[i],  # x-coordinate (district)
        value,  # y-coordinate (population exposed)
        f'{int(value)}',  # Text to display (the population exposed as an integer)
        fontsize=10, ha='center', va='bottom', color='black'  # Label styling
    )

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

# Adjust layout to ensure everything fits
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define the threshold for water extent variation (in km²)
threshold = 0.1 # Example: Only include data where the variation is greater than 100 km²

# Filter the data based on the threshold
filtered_data = df2[
    (df2['Minimum water extent variation between  [10-24 September 2024] & [24 September - 07 October 2024] (km2)'] > threshold) |
    (df2['Maximum water extent variation between  [10-24 September 2024] & [24 September - 07 October 2024] (km2)'] > threshold)
]

# Set up the figure
plt.figure(figsize=(12, 6))

# Plot the Minimum Water Extent Variation
plt.plot(
    filtered_data['District'],
    filtered_data['Minimum water extent variation between  [10-24 September 2024] & [24 September - 07 October 2024] (km2)'],
    marker='o', label='Minimum Water Extent Variation', color='blue'
)

# Plot the Maximum Water Extent Variation
plt.plot(
    filtered_data['District'],
    filtered_data['Maximum water extent variation between  [10-24 September 2024] & [24 September - 07 October 2024] (km2)'],
    marker='s', label='Maximum Water Extent Variation', color='red'
)

# Add titles and labels
plt.title('Water Extent Variation between [10-24 Sep. 2024] & [24-07 Oct. 2024]', fontsize=14)
plt.xlabel('District', fontsize=12)
plt.ylabel('Water Extent Variation (km²)', fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.grid(True)
plt.legend(fontsize=10)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()


```
